```python
#Problem 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0
    
    for like in likes_list:
        total_likes += like
    
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)

analyze_post_engagement([200, 300, 250, 300])
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
# Function to check rainfall adequacy
def rainfall_checker(rainfall_list, required_level):
    total = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total += rain
    
    average = total / len(rainfall_list)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_checker([60, 80, 75], 70)
```

```python
#Problem 4: Social Media – Duplicate Account Detection
# Function to detect duplicate usernames
def detect_duplicates(usernames):
    
    # Convert list to set to remove duplicates
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

detect_duplicates(["john", "alice", "john", "mark"])
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter
# Function to filter premium crop prices
def premium_crop_filter(prices):
    premium_crops = []
    
    # Filter prices greater than 2000
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)

premium_crop_filter([1500, 2500, 3200, 1800])
```

```python
#Problem 7: System Monitoring – Application Health Checker
# Function to check application health
def application_health(errors):
    
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)

application_health(7)
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker
# Function to check daily transaction limit
def transaction_checker(amount):
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_checker(60000)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System
# Function to calculate attendance and check eligibility
def attendance_checker(attendance_list):
    
    total_classes = len(attendance_list)
    attended = 0
    
    # Count attended classes
    for status in attendance_list:
        if status == 1:  # 1 means present
            attended += 1
    
    percentage = (attended / total_classes) * 100
    
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

attendance_checker([1,1,1,0,1,1,1,1,0,1])
```

```python
#Problem 10: Smart Electricity Bill Analyzer
# Function to calculate electricity bill
def electricity_bill(units):
    bill = 0
    
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    print("Total Bill:", bill)
    print("Usage Level:", status)

electricity_bill(250)
```

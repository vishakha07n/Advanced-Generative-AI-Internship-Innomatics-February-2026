```python
# Problem 1: Unique Words in a Sentence

# Taking input from user
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to remove duplicates
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)

```

```python
# Problem 2: Highest Salary from Employee Data

# Dictionary storing employee names and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
# Problem 3: Find Maximum and Minimum Values

# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:           # Loop through list
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Problem 4: Count Products Above 1000

# List of product prices
prices = [450, 1200, 899, 1500, 300]

count = 0  # Counter variable

# Loop to check prices
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

```python
# Problem 5: Calculate Attendance Percentage

# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Counting present days
present_count = attendance.count("P")

# Calculating percentage
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

# List of phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7: Count Character Frequency

# Given string
text = "python"

# Empty dictionary to store frequency
char_count = {}

# Loop through each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

```python
# Problem 8: Convert List to Tuple

# Given list
numbers = [10, 20, 30]

# Converting list to tuple
numbers_tuple = tuple(numbers)

print("Tuple:", numbers_tuple)
```

```python
# Problem 9: Check if Key Exists

# Dictionary of employees
employees = {
    "Ravi": 75000,
    "Anita": 68000
}

# Key to check
key_to_check = "Ravi"

# Checking if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10: Calculate Average Marks

# List of marks
marks = [70, 75, 80, 65]

# Calculating average
average = sum(marks) / len(marks)
print("Average Marks:", average)
```

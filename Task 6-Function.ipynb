```python
#Problem 1: Smart Parking Lot Management System

def smart_parking(capacity, logs):
    parked = 0
    peak = 0   # To track maximum parked vehicles at any time
    
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        # Track peak usage
        if parked > peak:
            peak = parked
    
    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)
    
    # Capacity check
    if parked > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, logs)
```

```python
#Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    # Base time (assume 5 minutes per km)
    time = distance * 5
    
    # Traffic impact
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8
    
    # Weather impact
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20
    
    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
#Problem 3: Movie Theatre Seat Occupancy Analyzer

def theatre_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    
    occupancy = (booked / total_seats) * 100
    
    print("Occupancy:", int(occupancy), "%")
    
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

theatre_occupancy(total_seats, booked_seats)
```

```python
#Problem 4: Cloud Server Load Classification System

def server_load(cpu_readings):
    total = 0
    
    # Calculate sum using loop
    for reading in cpu_readings:
        total += reading
    
    avg = total / len(cpu_readings)
    
    print("Average CPU Load:", int(avg), "%")
    
    if avg < 50:
        print("Server Status: Normal")
    elif 50 <= avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]
server_load(cpu_readings)
```

```python
#Problem 5: Smart Classroom Resource Usage Monitor

def classroom_monitor(resources):
    overused = []
    
    # Assume overuse if usage > 8 hours
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    
    print("Overused Resources:", ", ".join(overused) if overused else "None")
    
    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")


# Sample Input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_monitor(resources)
```

```python
#Problem 6: Online Event Registration Capacity Controller

def event_registration(capacity, registrations):
    
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```

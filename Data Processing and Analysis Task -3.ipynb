```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Identify employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display results
print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      f"(Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation (if any)
import string
query = query.translate(str.maketrans("", "", string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency of each word
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Step 5: Display words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Step 1: Loop through readings with index
for hour, reading in enumerate(sensor_readings):
    
    # Step 2: Check if reading is even (valid)
    if reading % 2 == 0:
        valid_readings.append((hour, reading))

# Step 3: Display valid readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Step 1: Extract domain and count occurrences
for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 2: Calculate percentage usage
total_emails = len(emails)

print("Email Domain Usage Percentage:")

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Detect spikes (more than 30% above average)
threshold = average_sales * 1.30

print("Sales Spike Days:")

for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Step 1: Count occurrences
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# Step 2: Display duplicates only
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```

```python
#Problem 1: Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display(self):
        print(self.name, "is", self.status)


name = input("Light Name: ")
action = input("Action: ")

light = SmartLight(name)

if action == "ON":
    light.turn_on()
else:
    light.turn_off()

light.display()
```

```python
#Problem 2: Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")
emp = Employee(name, emp_id, department)
emp.display()
```

```python
#Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
name = input("Contact Name: ")
phone = input("Phone Number: ")
c = Contact(name, phone)
c.display()
```

```python
#Problem 4: Product Price Tag Generator

class Product:
    
    # Constructor to initialize product name and price
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    # Method to display price tag
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))
product = Product("Headphones", 2499)
product.display_price_tag()
```

```python
#Problem 5: Movie Rating Display System

class Movie:
    
    # Constructor to initialize movie name and rating
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    # Method to display movie details
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")
movie = Movie("Inception", 4.8)
movie.display_movie()
```

```python
#Problem 6: Delivery Address Manager

class Delivery:
    # Constructor to initialize customer name and address
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    # Method to display delivery details
    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)
delivery = Delivery("Suman", "Hyderabad")
delivery.display_delivery()
```

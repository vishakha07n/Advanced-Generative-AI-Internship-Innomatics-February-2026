```python
#Problem 1: Mobile Recharge Validation System
# Function to validate recharge
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Recharge amount must be at least ₹50"
    elif amount not in valid_plans:
        return "Invalid plan selected"
    else:
        return "Recharge Successful"

# Retry using while loop
while True:
    amount = int(input("Enter recharge amount: "))
    result = validate_recharge(amount)
    print(result)

    if result == "Recharge Successful":
        break
    else:
        print("Please try again.\n")
```

```python
#Problem 2: Inventory Reorder Alert System
# Function to check inventory
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")

# Sample dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System
# Function to process result
def student_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example usage
marks_list = [60, 70, 45, 55]
result = student_result(marks_list)
print("Result:", result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
# Function to calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak == "yes":
        fare = fare + (fare * 0.25)  # 25% extra

    return fare

# Retry using while loop
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
#Problem 5: Employee Attendance Eligibility Checker

# Function to check eligibility
def attendance_check(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example usage
attendance = ["P", "P", "A", "P", "P", "A", "P"]
status = attendance_check(attendance)
print("Status:", status)
```

```python
#Problem 6: Password Strength Checker
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False

    # Check minimum length
    if len(password) < 8:
        return "Password too short"

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    # Check conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

# Example usage
pwd = input("Enter password: ")
print(check_password(pwd))
```
